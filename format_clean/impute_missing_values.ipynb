```python
import pandas as pd
```

```python
df = pd.read_pickle('sfcompo_nuc29_nulls.pkl')
```

```python
nuc29 = ['am241', 'am242m', 'am243', 'cm242', 'cm244', 
         'cs134', 'cs137', 'eu154', 'nd143', 'nd144', 
         'nd145', 'nd146', 'nd148', 'nd150', 'np237', 
         'pu238', 'pu239', 'pu240', 'pu241', 'pu242', 
         'sm147', 'sm149', 'sm150', 'sm151', 'sm152',
         'u234', 'u235', 'u236', 'u238'
        ]
```

```python
avg_vals = {}
for nuc in nuc29:
    avg_vals[nuc] = df[nuc].mean()
avg_vals
```

```python
for nuc in nuc29:
    df[nuc] = df[nuc].fillna(value=avg_vals[nuc])
```
